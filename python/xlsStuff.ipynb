```python
import pandas as pd

# Load the first Excel file
file1_path = "C:/Users/artor/Downloads/rcs/mgs010_Stampa Anagrafica Articoli al 21-08 con upa e giac.xls"  # Replace with the actual path "C:/Users/artor/Downloads/rcs/mgs010_Stampa Anagrafica Articoli al 21-08 con upa e giac.xls"
file2_path = "C:/Users/artor/Downloads/rcs/mgs010_Stampa Anagrafica Articoli al 21-08 con ubi e giac.xls"  # Replace with the actual path "C:/Users/artor/Downloads/rcs/mgs010_Stampa Anagrafica Articoli al 21-08 con ubi e giac.xls"

# Load all sheets from both files
df1 = pd.concat(pd.read_excel(file1_path, sheet_name=None), ignore_index=True)
df2 = pd.concat(pd.read_excel(file2_path, sheet_name=None), ignore_index=True)

print(df2.head)

output1 = "C:/Users/artor/Downloads/rcs/filewithprices.csv"  # Replace with your desired output path

output2 = "C:/Users/artor/Downloads/rcs/filewithlocation.csv"  # Replace with your desired output path

df1.to_csv(output1, index=False)
df2.to_csv(output2, index=False)

# Filter df2 where 'UBICAZIONE' contains "C.00.", case insensitive
df2_filtered = df2[df2["UBICAZIONE"].str.contains("C\.00\.", case=False, na=False)]

# Merge df1 and filtered df2 based on 'CODICE' and 'MARCA'
merged_df = pd.merge(
    df2_filtered[["CODICE", "MARCA"]],
    df1[["CODICE", "MARCA", "PRZ. ULT. ACQ."]],
    on=["CODICE", "MARCA"],
    how="inner",
)

# Save the result to a CSV file
output_path = (
    "C:/Users/artor/Downloads/rcs/output.csv"  # Replace with your desired output path
)
merged_df.to_csv(output_path, index=False)

print(f"Output saved to {output_path}")
```
